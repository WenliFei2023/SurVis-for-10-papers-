```python
import matplotlib.pyplot as plt
import numpy as np

categories = [
    "Traffic Flow Prediction",
    "Traffic Speed Prediction",
    "Arrival Time Prediction",
    "Overall Traffic Prediction"
]

papers = [
    ["Hussain et al. (2021)", "Lee et al. (2021)", "Pan et al. (2019)", "Tang et al. (2018)", "Zhou et al. (2022)"],
    ["Lin et al. (2017)", "Zhou et al. (2022)"],
    ["Abidin et al. (2014)", "Li et al. (2011)"],
    ["Bayati et al. (2018)", "Lee et al. (2021)", "Yin et al. (2021)"]
]

years = [
    [2021, 2021, 2019, 2018, 2022],
    [2017, 2022],
    [2014, 2011],
    [2018, 2021, 2021]
]

fig, ax = plt.subplots(figsize=(10, 6))

colors = plt.cm.tab10.colors[:len(papers)]

for i in range(len(categories)):
    ax.barh(np.arange(len(papers[i])), np.ones(len(papers[i])), height=0.6, left=i, color=colors[i % len(colors)])

    for j in range(len(papers[i])):
        ax.text(i, j, str(years[i][j]) + "\n" + papers[i][j], ha='center', va='center', color='white')

ax.set_yticks(np.arange(len(papers[0])))
ax.set_yticklabels([f"Paper {i+1}" for i in range(len(papers[0]))], fontsize=12)

ax.set_xticks(range(len(categories)))
ax.set_xticklabels(categories, rotation=45, ha='right', fontsize=12)

ax.set_xlabel('Prediction Objective', fontsize=14)
ax.set_ylabel('Papers', fontsize=14)
ax.set_title('Traffic Prediction Papers categoried by prediction objectives ', fontsize=16)

plt.tight_layout()
plt.show()
```
